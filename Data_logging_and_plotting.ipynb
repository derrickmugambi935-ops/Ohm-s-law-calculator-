```python
import random, time
import matplotlib.pyplot as plt

humidity = []

for t in range(10):
    hum = 50 + random.uniform(-5, 5)   # example: around 50% humidity
    humidity.append(hum)
    with open("humidity.csv", "a") as f:
        f.write(f"{t},{hum}\n")
    time.sleep(0.5)

plt.plot(humidity)
plt.xlabel("Time (s)")
plt.ylabel("Humidity (%)")
plt.title("Humidity Log")
plt.show()
```
